```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("../data/raw/multiple_choice_responses.csv", low_memory=False)

df
```

```python
# Does not have much meaning, other column with QX_OTHER_TEXT also same
df['Q2_OTHER_TEXT'].unique()
```

```python
# Drop columns that contain '_OTHER_TEXT' in their names
df = df.drop(columns=[col for col in df.columns if '_OTHER_TEXT' in col])
```

```python
# Loop over each question number and combine part columns into one column for each question
for i in range(1, 35):
    # Find all columns related to this question, like Q1_Part_1, Q1_Part_2, etc.
    part_columns = [col for col in df.columns if col.startswith(f"Q{i}_Part_") or col == f"Q{i}"]
    
    # Combine non-null entries for each question into a single string and save to a new column QX
    df[f'Q{i}'] = df[part_columns].apply(lambda row: ', '.join(row.dropna().astype(str)), axis=1)

# Drop the original part columns if they are no longer needed
df = df.drop(columns=[col for col in df.columns if "_Part_" in col or "_OTHER_TEXT" in col])
```

```python
# Drop the column: Time from Start to Finish (seconds)
# As this column has no much meaning for our study
df = df.drop('Time from Start to Finish (seconds)', axis=1)
```

```python
# Set the first row as column names
df.columns = df.iloc[0]

# Remove the first row of data
df = df.drop(index=0).reset_index(drop=True)
```

```python
df.columns = df.columns.str.replace('- Selected Choice', '', regex=False)
df.columns = df.columns.str.replace(r'\(Select all that apply\).*', '', regex=True)
df.columns = df.columns.str.replace(r'\(Include text response\).*', '', regex=True)
```
